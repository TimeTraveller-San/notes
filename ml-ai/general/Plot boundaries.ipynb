# Imports

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pandas as pd
from sklearn.metrics import f1_score
from sklearn.neural_network import MLPClassifier
```

# Load data

```python
# Loading iris data
iris = datasets.load_iris()
X = iris.data[:, [0, 2]]
y = iris.target
```

```python
# Make circles
def circle(r=10, n=100, label=0):
    points = []
    xs = np.linspace(-r, r, n//2)
    for x in xs:
        y = np.sqrt(abs(r**2 - x**2))
        points.append([x, y])
        points.append([x, -y])
    labels = np.full((n,), label)   
    return np.array(points), labels
```

```python
x1, labels1 = circle(10, 100, 0)
x2, labels2 = circle(30, 100, 1)
x3, labels3 = circle(50, 100, 2)
```

```python
X = np.concatenate((x1,x2,x3))
y = labels = np.concatenate((labels1,labels2,labels3))
```

# Model and plot

## Random Forest

```python
for est in range(1,10):
    model = RandomForestClassifier(n_estimators=est, max_depth=2)
    model.fit(X, y)
    y_preds = model.predict(X)
    score = f1_score(y, y_preds, average='micro')
    plt.figure(figsize=(5, 5))
    plot_decision_regions(X, y, clf=model, legend=2)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title(f'F1 score: {score}')
    plt.show()
```

## MLP classifier

```python
for layer_size in range(1, 10):
    model = MLPClassifier(hidden_layer_sizes=(layer_size,),learning_rate_init=0.1)
    model.fit(X, y)
    y_preds = model.predict(X)
    score = f1_score(y, y_preds, average='micro')
    plt.figure(figsize=(5, 5))
    plot_decision_regions(X, y, clf=model, legend=2)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title(f'F1 score: {score}')
    plt.show()
```
